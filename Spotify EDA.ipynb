# Exploratory Data Analysis on Spotify 2023 Dataset

This deliverable involves conducting an exploratory data analysis (EDA) on a dataset of popular tracks from Most Streamed Spotify Songs 2023. The goal is to analyze, visualize, and interpret the data to uncover meaningful insights.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
# import .csv file
spotify = pd.read_csv('spotify-2023.csv', encoding='latin1')
spotify
```

### Basic Descriptive

#### General Statistics of the Dataset

```python
spotify.describe()
```

#### What are the mean, median, and standard deviation of the streams column?

```python
# Coerce Errors/Non-numeric values to NaN
spotify['streams'] = pd.to_numeric(spotify['streams'], errors='coerce')
```

```python
# Main Codes to find descriptive statistics
sp_ave = spotify['streams'].mean()
sp_med = spotify['streams'].median()
sp_std = spotify['streams'].std()

# Display
sp_stats = pd.DataFrame({"Statistic": ["Average", "Median", "Standard Deviation"],"Value": [sp_ave, sp_med, sp_std]})
sp_stats.style
```

#### What is the distribution of released_year and artist_count? Are there any noticeable trends or outliers?

```python
# Plotting the distribution of released_year
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(spotify['released_year'], bins=15, kde=True, color='#afd7f6')
plt.title('Distribution of Release Year')
plt.xlabel('Release Year')
plt.ylabel('Count')

# Plotting the distribution of artist_count
plt.subplot(1, 2, 2)
sns.histplot(spotify['artist_count'], bins=10, kde=True, color='#b57fd3')
plt.title('Distribution of Artist Count')
plt.xlabel('Artist Count')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

> <p>The <b>Distribution of Release Year</b> shows a high increase in songs around 2000 onward, indicating the dataset favors recent releases.</p>
> <p>The <b>Distribution of Artist Count</b> highlights that most songs involve a single artist, with fewer collaborations as the number of artists increases.</p>
> <p><b>These trends reflect a dataset focused on modern music, primarily featuring solo performances.</b></p>

### Top Performers

#### Which track has the highest number of streams? Display the top 5 most streamed tracks.

```python
# Arrange the list of tracks according to the amount of streams
top_tr = spotify.sort_values(by='streams', ascending=False).reset_index(drop=True)

# Display track name, artist name, amount of streams
top_tr[['track_name', 'artist(s)_name', 'streams']].head()
```

#### Who are the top 5 most frequent artists based on the number of tracks in the dataset?

```python
# Separate Artists for those with more than 1 artist count
solo_artist = spotify['artist(s)_name'].str.split(', ').explode()
solo = solo_artist.str.strip()

# Count an artist's frequency then arrange from most to least
sp_artist = solo.value_counts().head()

# Display
sp_artist.name = 'Count'
sp_artist.index.name = 'artist(s)_name'
display(sp_artist.reset_index())
```

### Temporal Trends

#### Analyze the trends in the number of tracks released over time. Plot the number of tracks released per year.

```python
# Count how many songs are released in a given year
tpy = spotify['released_year'].value_counts().sort_index()

# Display
tpy.name = 'Count'
tpy.index.name = 'artist(s)_name'
display(tpy.reset_index())
```

```python
# Graphical Interpretation
plt.figure(figsize=(12, 6))
plt.plot(tpy.index, tpy.values, marker='o', color='#673ad5')
plt.title('Number of Tracks Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Tracks')
plt.grid(True)
plt.show()
```

> Most songs on the list were released from <b>2021 to 2023</b>

#### Does the number of tracks released per month follow any noticeable patterns? Which month sees the most releases?

```python
# Count how many songs were released each month
tpm = spotify['released_month'].value_counts().sort_index()

# Convert months' numeric values to their names
moname = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June',
          7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}
months = tpm.index.map(moname)

# Display Tracks released per Month
tpm.index = months
tpm.index.name = 'Month'
tpm.name = 'Tracks Per Month'
display(tpm.reset_index(drop=False))
```

```python
# Graphical Interpretation
colors = sns.color_palette('BuPu', 4)
plt.figure(figsize=(12, 6))
plt.bar(months, tpm, color=colors)
plt.title('Number of Tracks Released Per Month')
plt.xlabel('Month')
plt.ylabel('Number of Tracks')
plt.grid(axis='y')
plt.xticks(rotation=45)
plt.show()
```

> <p>Based on the graph, there is a <b>higher number of song releases in the first half</b> of the year. There is also a <b>gradual increase of song releases in the last quarter</b> of the year</p>
> <p>Most songs are released on <b>January</b>, closely followed by <b>May</b>.</p>

### Genre and Music Characteristics

#### Examine the correlation between streams and musical attributes <b><u>like</u></b> bpm, danceability_%, and energy_%. Which attributes seem to influence streams the most?

```python
# Author highlighted "like" to indicate that bpm, danceability_%, and energy_% will not be the only ones that will be correlated to the streams count
```

```python
# Correlate streams and musical attributes to each other
mus_att = spotify[['streams', 'bpm', 'danceability_%', 'energy_%', 'valence_%', 'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%']].corr()

# Display correlation of streams to musical attributes
corr = pd.DataFrame({'Correlation to Streams' : mus_att['streams']})
corr
```

> Overall, nearly all musical attributes exhibit <b>weak or insignificant negative correlations</b> with the number of streams, whereas <b>none exhibit high connections</b>. This implies that <b><u>the number of streams in the dataset is not substantially impacted by variables like tempo (bpm), danceability, or energy.</u></b>

```python
# Graphical Interpretation of Correlation of Streams and Musical Attributes to each other
plt.figure(figsize=(12, 6))
sns.heatmap(mus_att, annot=True, cmap='BuPu', linewidths=0.5)
plt.title('Correlation Between Streams and Musical Attributes')
plt.xticks(rotation=30)
plt.show()
```

> According to the heat map, <b>streams have the most correlation with danceability_% and enerygy_%</b>.
> <br>Furthermore, <b>danceability_% and valence_%</b> have the highest correlation with each other, as well as <b>energy_% and valence_%</b>

#### Is there a correlation between danceability_% and energy_%? How about valence_% and acousticness_%?

```python
# Correlate danceability_% and energy_%
dance_energy = spotify['danceability_%'].corr(spotify['energy_%'])
print(f'Correlation between Danceability and Energy: {dance_energy}')

# Display valence_%', acousticness_%
val_ac = spotify['valence_%'].corr(spotify['acousticness_%'])
print(f'Correlation between Valence and Acousticness: {val_ac}')

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Display Scatterplot
axes[0].scatter(x='danceability_%', y='energy_%', data=spotify, color='#9071C5')
axes[0].set_title("Danceability vs Energy")
axes[0].set_xlabel('Danceability (%)')
axes[0].set_ylabel('Energy (%)')

# Display Scatterplot
axes[1].scatter(x='valence_%', y='acousticness_%', data=spotify, color='#7186C5')
axes[1].set_title('Valence vs Acousticness')
axes[1].set_xlabel('Valence (%)')
axes[1].set_ylabel('Acousticness (%)')

plt.tight_layout()
plt.show()
```

> Between <b>danceability_% and energy_%</b>, there is <b>a weak positive correlation</b> suggests that, in general, songs that are easier to dance to tend to have slightly higher energy, but the relationship is not very strong.
> <br>For <b>valence_% and acousticness_%</b>, the <b>small negative correlation</b> indicates that songs with a higher mood (valence) tend to have a very slight tendency to be less acoustic, but the effect is weak and not very significant.

### Advanced Analysis


#### Based on the streams data, can you identify any patterns among tracks with the same key or mode (Major vs. Minor)?

##### Grouped by Key

```python
# Count the number of songs that used the same key
sp_key = spotify.groupby('key').size()

# Display
sp_key.name = 'Count'
sp_key.index.name = 'Key'
display(sp_key.reset_index())
```

```python
# Graphical Interpretation
colors = sns.color_palette('BuPu', 4)
plt.figure(figsize=(12, 5))
plt.bar(sp_key.index, sp_key.values, color=colors)
plt.title('Streams by Key')
plt.xlabel('Key')
plt.ylabel('Streams')
plt.show()
```

> Based on the data and graph, <b>C# is frequently used</b>. Despite that, the balanced distribution of keys indicates that <b>artists exploreÂ with diverse tonalities</b>. The dynamic and varied character of current music is further <b>enhanced by this variability in key selection</b>.

##### Grouped by Mode

```python
# Count the number of songs that used the same mode
sp_mode = spotify.groupby('mode').size()

# Display
sp_mode.name = 'Count'
sp_mode.index.name = 'Mode'
display(sp_mode.reset_index())
```

```python
# Graphical Interpretation
colors = sns.color_palette('BuPu', 2)
plt.figure(figsize=(12, 5))
plt.bar(sp_mode.index, sp_mode.values, color=colors)
plt.title('Streams by Mode')
plt.xlabel('Mode')
plt.ylabel('Streams')
plt.show()
```

> The data suggests that artists prefer the Major mode slightly more frequently than the Minor mode, as <b>there are more songs composed in Major</b>. However, the difference is not very large, indicating that both modes are commonly used and important in musical composition.

##### Grouped by the same Key AND Mode

```python
# Count the number of songs that used the same key AND mode
sp_keymode = spotify.groupby(['key', 'mode']).size()

# Display
sp_keymode.name = 'Count'
sp_keymode.index.name = 'Key and Mode'
display(sp_keymode.reset_index())
```

```python
# Graphical Interpretation
sp_km = sp_keymode.reset_index()
plt.figure(figsize=(12, 6))
colors = sns.color_palette('BuPu', 2)
sns.barplot(x='key', y='Count', hue='mode', data=sp_km, palette=colors)
plt.title('Number of Songs by Key and Mode')
plt.xlabel('Key')
plt.ylabel('Count')
plt.legend(title='Mode')
plt.grid(axis='y')
plt.show()
```

> <b>C# Major, D Major, G Major, and G# Major are the ones used most often in songs</b>. However, besides those four keys, <b>minor keys tend to dominate</b> in this dataset. This could reflect a trend in music where <b>songs in minor keys are more common than those in major keys</b>, or it could be influenced by the types of artists or genres surveyed.

#### Do certain genres or artists consistently appear in more playlists or charts? Perform an analysis to compare the most frequently appearing artists in playlists or charts.

```python
# Separate Artists for those with more than 1 artist count
spotify['artist(s)_name'] = spotify['artist(s)_name'].str.split(', ')
spotify = spotify.explode('artist(s)_name')
spotify['artist(s)_name'] = spotify['artist(s)_name'].str.strip()

# Coerce/Fix errors and non-numerical values in charts and playlists
spotify['in_apple_charts'] = pd.to_numeric(spotify['in_apple_charts'], errors='coerce')
spotify['in_deezer_playlists'] = pd.to_numeric(spotify['in_deezer_playlists'].astype(str).str.replace(',', ''), errors='coerce')
spotify['in_shazam_charts'] = pd.to_numeric(spotify['in_shazam_charts'].astype(str).str.replace(',', ''), errors='coerce').fillna(0).astype(int)

# Count sum of artists' popularity based on charts and playlists
artist = spotify.groupby('artist(s)_name')[['in_spotify_playlists', 'in_apple_playlists', 'in_deezer_playlists', 'in_spotify_charts', 'in_apple_charts', 'in_deezer_charts', 'in_shazam_charts']].sum()
artist['total_mentions'] = artist.sum(axis=1)

# Display Top 10
top_art = artist.sort_values(by='total_mentions', ascending=False).head(10)

top_art.style
```

> <p><b>The Weeknd</b> has the most appearances through the playlists and charts.
    <br>The Weeknd also has the highest appearance in the <b>Spotify Playlists, Apple Playlists and Charts, and Shazam Charts</b>.</p>
> <p><b>Eminem</b> has the most appearances in <b>Deezer Playlists</b>.
    <br><b>Bad Bunny</b> has the most appearances in <b>Spotify and Deezer Charts.</b></p>

```python
# Graphical Interpretation
plt.figure(figsize=(10,6))
sns.barplot(x=top_art.index, y=top_art['total_mentions'], hue=top_art.index, palette='BuPu', edgecolor="black", linewidth=0.25)
plt.xlabel('Artist')
plt.ylabel('Total Mentions')
plt.title('Top 10 Artists by Total Mentions Across Platforms')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Playlist Summary
art_play = top_art[['in_spotify_playlists', 'in_apple_playlists', 'in_deezer_playlists']]

# Plot the data
colors = sns.color_palette('BuPu', 3)
art_play.plot(kind='bar', figsize=(12, 6), color=colors)
plt.title('Top 10 Artists by Mentions in Playlists (Spotify, Apple, Deezer)')
plt.xlabel('Artists')
plt.ylabel('Mentions')
plt.xticks(rotation=45, ha="right")
plt.legend(title="Platforms")
plt.tight_layout()
plt.show()
```

```python
# Charts Summary
art_char = top_art[['in_spotify_charts', 'in_apple_charts', 'in_deezer_charts', 'in_shazam_charts']]

# Plot the data
colors = sns.color_palette('BuPu', 4)
art_char.plot(kind='bar', figsize=(12, 6), color=colors)
plt.title('Top 10 Artists by Mentions in Charts (Spotify, Apple, Deezer, Shazam)')
plt.xlabel('Artists')
plt.ylabel('Mentions')
plt.xticks(rotation=45, ha="right")
plt.legend(title="Platforms")
plt.tight_layout()
plt.show()
```
